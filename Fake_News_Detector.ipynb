## Project: Fake News Detector

```python
# Importing Libraries
import numpy as np # Used for Numerical Operations
import pandas as pd # Used for Data Manipulation and Analysis(Data Cleaning)
import re # Used for Regular Expression Operations(Because we are dealing with text data)
from nltk.corpus import stopwords # Used for removing stopwords like 'the', 'is', 'in', 'with', 'for' etc...
from nltk.stem.porter import PorterStemmer # Used to reduce words to their base form e.g: Loving/Loved == Love
from sklearn.feature_extraction.text import TfidfVectorizer # Used to convert words into vector form e.g: Love == [0.0]
from sklearn.model_selection import train_test_split # Used to split dataset into training and testing data sets
from sklearn.linear_model import LogisticRegression # Used for creating and fitting model
from sklearn.metrics import accuracy_score # Used to calculate accuracy of model 
```

```python
# Importing/Loading dataset
df = pd.read_csv('WELFake_Dataset.csv')
df.head()
```

```python
# Checking shape of dataset
df.shape
```

```python
# Checking Null values
df.isna().sum()
```

```python
# Filling Null values with an empty space
df = df.fillna(' ')
```

```python
# look now there are no Null values in our Dataset
df.isna().sum() 
```

```python
# Stemming
ps = PorterStemmer()
def stemming(title):
    stemmed_title = re.sub('[^a-zA-Z]', " ", title)
    stemmed_title = stemmed_title.lower()
    stemmed_title = stemmed_title.split()
    stemmed_title = [ps.stem(word)for word in stemmed_title if not word in stopwords.words('english')]
    stemmed_title = " ".join(stemmed_title)
    return stemmed_title
```

```python
import nltk
nltk.download('stopwords', quiet=True)  # Suppresses extra output
```

```python
df['title'] = df['title'].apply(stemming)
df['title']
```

```python
# Splitting Dataset into X and y variables  
x = df['title'].values # Input
y = df['label'].values # Output
```

```python
# Converting values of X in vector form so our machine learning model can understand it
vector = TfidfVectorizer()
x = vector.fit_transform(x)  # Fitting and transforming in one step
```

```python
# Printing 'x' to check if it's been updated
print(x)
```

```python
# Splitting dataset into training and testing sets (80/20) ratio 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2) 
```

```python
# Checking shape of x_train
x_train.shape
```

```python
# Checking shape of x_test
x_test.shape 
```

```python
# Creating and Fitting our model 
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100).fit(x, y)
model
```

```python
# Checking our model accuracy on the y_train data
train_y_pred = model.predict(x_train)
print("Training Accuracy:",accuracy_score(train_y_pred, y_train)*100,'%')
```

```python
# Checking our model accuracy on the y_test data
test_y_pred = model.predict(x_test)
print("Testing Accuracy:",accuracy_score(test_y_pred, y_test)*100,"%")
```

```python
# Prediction System
input_data = x_test[5]
prediction = model.predict(input_data)
if prediction[0] == 1:
    print("Fake news")
else:
    print("Real news") 
   
```
